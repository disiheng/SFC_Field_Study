```python
import os
import sys

fpath = "/Users/liming/Nutstore/SFC_Field_Study/mracs_cpp/mracs/wavelets"#/binary"

order = 7
fbase = "Daubechies" + "G"+str(order) + "Phi" + ".bin"

fname = os.path.join(fpath, fbase)
print(fname)
```

```python
import numpy as np

phiStart   = 0
phiEnd     = 2*order - 1
phiSupport = phiEnd - phiStart
SampRate   = 10000

sampOff = np.empty(phiSupport, dtype=np.int32)
for isamp in range(0, phiSupport):
    sampOff[isamp] = isamp * SampRate;
        
nsamp = SampRate * phiSupport
print(nsamp)

Daubechies_Phi = np.empty(nsamp, dtype=np.float64)
samp = np.arange(0, SampRate)
```

```python
if (not os.path.exists(fname)):
    print(("file not found:", fname))
    sys.exit()
    
f = open(fname,'rb')
Daubechies_Phi = np.fromfile(f, dtype=np.float64, count=nsamp)
f.close()
```
